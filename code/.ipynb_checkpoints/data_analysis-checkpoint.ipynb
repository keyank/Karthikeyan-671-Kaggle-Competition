## Frequency Analysis

```python
ordered_feature_list = [
'Weather',
'Temperature',
'Time',
'Coupon_validity',
'Gender',
'Age',
'Children',
'Bar',
'Coffeehouse',
'Carryaway',
'Restaurantlessthan20',
'Restaurant20to50',
'Direction_same',
'Distance', 
'Driving_to',
'Passanger',
'Income',
'Coupon',
'Maritalstatus',
'Education',
'Occupation',
]
```

```python
data = data.dropna()

def get_stats(feature, data): 
    column = data[feature]
    counts = list(dict(Counter(column)).items())
    counts.sort(key=lambda x: x[1], reverse=True)
    
    x = [str(item[0]) for item in counts]
    y = [item[1] for item in counts]
    return x, y



fig, axs =  plt.subplots(nrows=3, ncols=7, figsize=(60,30))

count = 0 

for i, feature in enumerate(ordered_feature_list) : 

    x, y = get_stats(feature, data)
    b = i%7
    a = math.floor(i/7)
    axs[a][b].bar(x, y)
    if a == 2: 
        axs[a][b].set_xticklabels(x, rotation=90, fontdict = {'fontsize' : 20})
    else: 
        axs[a][b].set_xticklabels(x, rotation=30, fontdict = {'fontsize' : 25})
    axs[a][b].set_ylabel('Counts')
    axs[a][b].set_title(feature, fontdict = {'fontsize' : 40})


fig.tight_layout(pad=2.5)

plt.show()
```

# Algorithmic Reliance:


```python
X_df, y_df = clean_all(data, drop_first, prune)

data_df = X_df.copy()
data_df['Decision'] = y_df

data_df = data_df.sample(frac=1, random_state=0)

num_train = int(0.7*len(data_df))

data_train = data_df[:num_train]
data_test = data_df[num_train:]


X_train_df, y_train_df = data_train[data_train.columns[:-1]], data_train[data_train.columns[-1]]
X_test_df, y_test_df = data_test[data_test.columns[:-1]], data_test[data_test.columns[-1]]

y_train = np.array(y_train_df)
y_test = np.array(y_test_df)

all_accuracies = {}
```

```python
X_train = np.array(X_train_df)
X_test = np.array(X_test_df)

(n_est, criterion, max_features, oob_score) = (200, 'entropy', 0.2, True)

avg_full_acc = []

for random_state in range(5): 
    clf = RandomForestClassifier(n_estimators=n_est, criterion=criterion, n_jobs=-1, 
                                                   max_features=max_features, oob_score=oob_score, 
                                 random_state=random_state)

    clf.fit(X_train, y_train)
    full_acc = clf.score(X_test, y_test)
    avg_full_acc.append(full_acc)
    

all_accuracies['all_features'] = np.mean(full_acc)
```

```python
initial_features = list(set([item.split(':')[0] for item in list(X_train_df.columns)]))

processed_features = list(X_train_df.columns)


for feature in initial_features: 
    
    avg_removed_acc = []
    
    for random_state in range(5):  
        
        X_train_df_removed =  X_train_df[[item for item in processed_features if item.split(':')[0] != feature]]
        X_test_df_removed =  X_test_df[[item for item in processed_features if item.split(':')[0] != feature]]
        X_train_removed = np.array(X_train_df_removed)
        clf = RandomForestClassifier(n_estimators=n_est, criterion=criterion, n_jobs=-1, 
                                                   max_features=max_features, oob_score=oob_score, 
                                 random_state=random_state)
        clf.fit(X_train_removed, y_train)
        removed_acc = clf.score(X_test_df_removed, y_test)
        avg_removed_acc.append(removed_acc)

    all_accuracies[feature] = np.mean(avg_removed_acc)
```

```python
all_accuracies 

algorithmic_reliance = {}

for k, v in all_accuracies.items(): 
    if k == 'all_features':
        continue
    else: 
        ar = (1-all_accuracies[k])/(1-all_accuracies['all_features'])
        algorithmic_reliance[k] = ar
        
```

```python
algorithmic_reliance
```

```python
results = list(algorithmic_reliance.items())

results.sort(key=lambda x: x[1] , reverse=True)

init_features = [item[0] for item in results]
ars = [item[1] for item in results]



fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(20,5))

axes.bar(init_features, ars)
axes.plot( init_features , [1]*len(init_features), color='red', linewidth=3, label='AR = 1')

axes.set_xticklabels(init_features, rotation=90)
axes.set_ylabel('Algorithmic Reliance')
axes.set_ylim([0.9, 1.3])


fig.suptitle('Algorithmic Reliance of all the features (Averaged across 5 seeds)', fontsize=15)
plt.legend()
```
