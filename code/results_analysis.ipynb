```python
rf_accs = pickle.load(open('../data/coupons/all_outputs/randomforest_accs_prob.pk', 'rb+'))
xgb_accs = pickle.load(open('../data/coupons/all_outputs/xgboost_accs_prob.pk', 'rb+'))
snn_accs = pickle.load(open('../data/coupons/all_outputs/simple_nn_accs_prob.pk', 'rb+'))
snn_accs = [v for k, v in snn_accs.items()]

random_nn_accs = pickle.load(open('../data/coupons/all_outputs/random_nn_accs_prob.pk', 'rb+'))
random_nn_accs = [v for k, v in random_nn_accs.items()]
```

```python
fig, axs =  plt.subplots(nrows=1, ncols=4, figsize=(25,5))

titles = ['Random Forest', 'XGBoost', 'Simple NN', 'Random NN']


for i, accs in enumerate([rf_accs, xgb_accs, snn_accs, random_nn_accs ]): 
    counts, vals, _ = axs[i].hist(accs, bins= int(np.sqrt(len(accs))) , histtype='bar', rwidth=0.9)
    axs[i].set_title(titles[i],  fontdict = {'fontsize' : 20})
    x_mean = [np.mean(accs), np.mean(accs)]
    x_max = [np.max(accs), np.max(accs)]
    x_min = [np.min(accs), np.min(accs)]
    y = [0, np.max(counts)]
    
    axs[i].plot(x_mean, y, linewidth=8, label='Mean:{}'.format(np.around(x_mean[0] , 3) ), color='darkorange')
    axs[i].plot(x_max, y, linewidth=5, label='Max:{}'.format(np.around(x_max[0] , 3) ), color='green')
    axs[i].plot(x_min, y, linewidth=5, label='Min:{}'.format(np.around(x_min[0] , 3) ), color='red')
    
    axs[i].legend(bbox_to_anchor=(1, -0.1), ncol=3)
    

plt.show()
```

```python
from data_processing import *

train_path = '../data/coupons/train.csv'
test_path = '../data/coupons/test.csv'
sample_path = '../data/coupons/sample_submission1.csv'

data = pd.read_csv(train_path)
test_data = pd.read_csv(test_path)

drop_first = False
prune=False

X_df, y_df = clean_all(data, drop_first, prune)
X = np.array(X_df)
y = np.array(y_df)

X.shape, y.shape
```

```python
defaults = {
    'n_est': 100, 
    'criterion': 'entropy', 
    'max_features': 0.2, 
    'oob_score': True
}


n_est_list = [10, 20, 50, 100, 200, 300, 500, 1000]
max_feature_list = [0.05, 0.1, 0.2, 0.3, 0.5, 0.7, 1]
```

```python
n_splits = 5

kf = KFold(n_splits=n_splits, shuffle=True)



n_est_accs = []

for i, (n_est) in enumerate(n_est_list): 
    
    avg_score = 0

    for train_index, val_index in kf.split(X):
        X_train, X_val = X[train_index], X[val_index]
        y_train, y_val = y[train_index], y[val_index]

        clf = RandomForestClassifier(n_estimators=n_est, n_jobs=-1)

        clf.fit(X_train, y_train)
        score = clf.score(X_val, y_val)
        avg_score += score 

    avg_score = avg_score/n_splits
    n_est_accs.append(avg_score)
    
```

```python
plt.plot(n_est_list, n_est_accs, 'o-')
```

```python
n_splits = 5

kf = KFold(n_splits=n_splits, shuffle=True)

max_feature_list = [0.05, 0.1, 0.2, 0.3, 0.5, 0.7, 1]


max_feature_accs = []

for i, (max_feature) in enumerate(max_feature_list): 
    
    avg_score = 0

    for train_index, val_index in kf.split(X):
        X_train, X_val = X[train_index], X[val_index]
        y_train, y_val = y[train_index], y[val_index]

        clf = RandomForestClassifier(n_jobs=-1, max_features=max_feature)

        clf.fit(X_train, y_train)
        score = clf.score(X_val, y_val)
        avg_score += score 

    avg_score = avg_score/n_splits
    max_feature_accs.append(avg_score)
    
```

```python
plt.plot(max_feature_list, max_feature_accs, 'o-')
```
